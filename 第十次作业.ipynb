1、用python分别绘制均值为0，方差为1、2、5、10的正态分布曲线，并采用μ=0,σ=2的正态分布生成一千个随机数，并绘制频率直方图

```python
#Python绘制正态分布曲线
import numpy as np
import matplotlib.pyplot as plt

def normpdf(x,mu,sigma):       
    pdf=np.exp(-(x-mu)**2/(2*sigma**2))/(sigma * np.sqrt(2 * np.pi))
    return pdf
i=0
color=["r","b","g","y"]
for i in range(0,4):
    mu,sigma=eval(input("输入 μ，σ="))  
    x= np.arange(mu-3*sigma,mu+3*sigma,0.01) 
    y=normpdf(x,mu,sigma)
    
    plt.plot(x,y,color=color[i],linewidth=2,label=str(mu)+str(sigma))
    plt.vlines(mu, 0, normpdf(mu,mu,sigma), colors = color[i], linestyles = "dotted")
    plt.vlines(mu+sigma, 0, normpdf(mu+sigma,mu,sigma), colors = color[i], linestyles = "dotted")
    plt.vlines(mu-sigma, 0, normpdf(mu-sigma,mu,sigma), colors = color[i], linestyles = "dotted")
    plt.xticks ([mu-sigma,mu,mu+sigma],['μ-σ','μ','μ+σ'])

    plt.grid()
   
```

```python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
mu,sigma=eval(input("输入 μ，σ="))
a=stats.norm.rvs(mu,sigma,size=1000)
print(a)
sns.set(style="darkgrid")
sns.distplot(a,hist=True,kde=False)
plt.show()
```

2、用极坐标绘制笛卡尔心形图

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,10000)
a=1
r=a*(1-np.cos(x))
plt.subplot(polar=True)
plt.plot(x,r,c='r')
plt.show()
```
