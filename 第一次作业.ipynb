# （2）、GPA的计算

```python
while True:
    try:
        x = int(input('输入成绩：'))
        if x<=100 and x>=90:
            print('成绩评级：A GPA：4.0')
        elif x<90 and x>=80:
            print('成绩评级：B GPA：3.0')
        elif x<80 and x>=70:
            print('成绩评级：C GPA：2.0')
        elif x<70 and x>=60:
            print('成绩评级：D GPA：1.0')
        elif x<60 and x>=0:
            print('成绩评级：E GPA：0.0')
        else: 
            print("成绩输入错误")
            break
    except ValueError:
        print('输入内容非数字或非整数')
        break
```

# 4、用递归与迭代法求n！，并比较两种算法对同一数字计算的运行时间
# （1）、迭代算法

```python
import time
n=int(input('输入一个正整数：'))
x=1
start_time=time.time()
for i in range(1,n+1):
    x=x*i
end_time=time.time()
run_time=end_time-start_time
print('n!=',x)
print('运行时间为',run_time,'s')
```

# （2）、递归算法

```python
import time
def f(x):
    if x==0 or x==1:
        return 1
    else:
        return (x*f(x-1))
n=int(input('输入一个正整数：'))
start_time=time.time()
x=f(n)
end_time=time.time()
run_time=end_time-start_time
print('n!=',x)
print('运行时间为',run_time,'s')
```

# 5、用户登录界面

# （1）、可输入五次

```python
user="abcd"
key="a121212"
for i in range(0,6):
    x=input('用户名')
    y=input('密码')
    if x==user and y==key:
        print('账号密码正确')
        break
    else:
        i+=1
        print('账号密码不匹配')
        if i==5:
            print('account locked')
            break
```

# （2）、可输入无数次

```python
user="abcd"
key="a121212"
while True:
    x=input('用户名')
    y=input('密码')
    if x==user and y==key:
        print('账号密码正确')
        break
    else:
        print('账号密码不匹配')
```
